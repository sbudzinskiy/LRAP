```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

### New: Fixed n

```python
def read_eps_varying_r(n, R, method, trials, folder=""):
    eps = np.zeros([len(R), len(trials)])
    for j in range(len(R)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{R[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_n(ax, n, R, method, ntrials, color, lw, ls, label, folder=""):
    eps = read_eps_varying_r(n, R, method, ntrials, folder)
    eps_mean = np.mean(eps, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(R, eps_mean, marker="o", color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(R, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(R, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
R1600 = [20, 40, 80, 160, 320, 640]
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plot_n(ax, 1600, R1600, "svd", np.arange(1,11), color=oi_colors["blue"], lw=3, ls="-", label=r"$n = 1600$")

x1600 = np.linspace(25, 640, 100)
ax.semilogy(x1600, 7e-1/np.sqrt(x1600), color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1/2})$")
ax.semilogy(x1600, 20/x1600, color=oi_colors["orange"], lw=3, ls="--", label=r"$O(r^{-1})$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
# ax.yaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax.xaxis.set_major_locator(plt.FixedLocator([20, 320, 640]))
# ax.set_xlim(-50, 2000)
ax.set_ylim(1e-2, 1.0)

# Titles
ax.set_xlabel(r"rank: $r$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='upper right', frameon=True, fontsize=fontsize, ncol=1)

# ax.set_rasterized(True)
# plt.savefig(f"./pics/randort_varying_n_rsvd.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/randort_varying_n_rsvd.eps", format='eps', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/mettx_identity_r.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```

### New: Fixed r

```python
def read_eps_varying_n(r, N, method, trials, folder=""):
    eps = np.zeros([len(N), len(trials)])
    for j in range(len(N)):
        for i in range(len(trials)):
            with open(f'./{folder}n{N[j]}_r{r}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_r(ax, r, N, method, trials, marker, color, lw, ls, label, folder=""):
    eps = read_eps_varying_n(r, N, method, trials, folder)
    eps_mean = np.mean(eps, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(N, eps_mean, marker=marker, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(N, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(N, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
N40 = [100, 150, 200, 250, 300, 350, 400, 450, 500, 600, 700, 800, 900, 1000, 1500, 2000, 2500, 3000, 4000, 5000, 10000, 20000] # 50
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plot_r(ax, 40, N40, "rsvd_p5_q1", np.arange(1,6), marker="o", color=oi_colors["bluegreen"], lw=3, ls="-", label=r"$r = 40$")

y40 = np.linspace(50, 20000, 1000)
ax.semilogy(y40, 4.5e-1*np.log(y40)/np.sqrt(y40), color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(\log(n) / \sqrt{n})$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
# ax.yaxis.set_major_locator(plt.MaxNLocator(5))
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax.xaxis.set_major_locator(plt.FixedLocator([50, 2500, 10000, 20000]))
# ax.set_xlim(-500, 10500)
ax.set_ylim(1e-2, 5e-1)
 
# Titles
ax.set_xlabel(r"size: $n$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='lower right', frameon=True, fontsize=fontsize, ncol=1)

# ax.set_rasterized(True)
# plt.savefig(f"./pics/randort_varying_n_rsvd.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/randort_varying_n_rsvd.eps", format='eps', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/mettx_identity_n.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```

### Fixed r

```python
r = 40
N_svd  = [120, 160, 200, 240, 280, 320, 360, 400, 440, 480, 520, 560, 600, 640, 680, 720, 760, 800, 840, 880, 920, 960, 1000, 1200, 1400, 1600, 1800]
N_rsvd = [1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 5000, 10000, 20000]
```

```python
def read_eps_varying_n(r, N, method):
    eps = np.zeros([len(N), 10])
    for j in range(len(N)):
        for i in range(10):
            with open(f'./n{N[j]}_r{r}_{method}/{i+1}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps
```

```python
eps_svd = read_eps_varying_n(r, N_svd, "svd")
eps_svd50 = np.percentile(eps_svd, 50, axis=1)
eps_svd25 = np.percentile(eps_svd, 25, axis=1)
eps_svd75 = np.percentile(eps_svd, 75, axis=1)
eps_svd10 = np.percentile(eps_svd, 10, axis=1)
eps_svd90 = np.percentile(eps_svd, 90, axis=1)

eps_rsvd = read_eps_varying_n(r, N_rsvd, "rsvd_p5_q1")
eps_rsvd50 = np.percentile(eps_rsvd, 50, axis=1)
eps_rsvd25 = np.percentile(eps_rsvd, 25, axis=1)
eps_rsvd75 = np.percentile(eps_rsvd, 75, axis=1)
eps_rsvd10 = np.percentile(eps_rsvd, 10, axis=1)
eps_rsvd90 = np.percentile(eps_rsvd, 90, axis=1)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plt.plot(N_svd, eps_svd50, marker="o", color=oi_colors["blue"], lw=3, ls="-", label="AP-SVD")
ax.fill_between(N_svd, eps_svd25, eps_svd75, color=oi_colors["blue"], lw=0, alpha=0.5)
ax.fill_between(N_svd, eps_svd10, eps_svd90, color=oi_colors["blue"], lw=0, alpha=0.3)
plt.plot(N_svd, 0.45*np.sqrt(np.log(N_svd)**2 / N_svd), color=oi_colors["orange"], lw=3, ls="--", label=r"$0.45 \cdot \sqrt{\log^2(n) / n}$")
# plt.plot(N_svd, 2*np.sqrt(np.log(N_svd) / N_svd), color=oi_colors["skyblue"], lw=3, ls="--", label=r"$0.45 \cdot \sqrt{\log(n) / n}$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.MaxNLocator(3))
ax.xaxis.set_major_locator(plt.FixedLocator([120, 1800]))
ax.set_xlim(20, 1900)
ax.set_ylim(6e-2, 1.05e-1)

# Titles
# ax.set_title(r"Absolute value of the smallest negative entry of $X_r$", fontsize=fontsize)
ax.set_xlabel(r"size: $n$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='lower left', frameon=True, fontsize=fontsize, ncol=1)

# plt.savefig(f"./pics/randort_varying_n_svd.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/randort_varying_n_svd.eps", format='eps', dpi=300, bbox_inches='tight', pad_inches=0)
```
