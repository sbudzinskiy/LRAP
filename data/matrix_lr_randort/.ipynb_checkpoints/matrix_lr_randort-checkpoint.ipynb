```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

```python
def lr_randort(n, r):
    U = np.random.default_rng().normal(0, 1, [n,r])
    U, _ = np.linalg.qr(U, mode='reduced')
    V = np.random.default_rng().normal(0, 1, [n,r])
    V, _ = np.linalg.qr(V, mode='reduced')
    A = U @ V.T
    return A

def norm_lr_randort(n, r, ntrials):
    norms = np.zeros(ntrials, dtype='float')
    for i in range(ntrials):
        X = lr_randort(n, r)
        X /= np.max(np.abs(X))
        norms[i] = np.linalg.norm(X, 2)
    return norms

def norm_lr_randort_stats(n, R, ntrials):
    nrms = np.zeros([len(R),ntrials], dtype='float')
    for i in range(len(R)):
        nrms[i] = norm_lr_randort(n, R[i], ntrials)
    return nrms
```

```python
def plot_norms(ax, R, nrms, marker, color, lw, ls, label):   
    nrms_target = np.percentile(nrms, 50, axis=1)
    nrms_50 = np.percentile(nrms, 50, axis=1)
    nrms_25 = np.percentile(nrms, 25, axis=1)
    nrms_75 = np.percentile(nrms, 75, axis=1)
    nrms_10 = np.percentile(nrms, 10, axis=1)
    nrms_90 = np.percentile(nrms, 90, axis=1)
    
    ax.semilogy(R, nrms_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(R, nrms_25, nrms_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(R, nrms_10, nrms_90, color=color, lw=0, alpha=0.3)
```

### Fixed n and r

```python
def read_eps_varying_r(n, r, K, method, trials, folder=""):
    eps = np.zeros([len(K), len(trials)])
    for j in range(len(K)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{r}_k{K[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_k(ax, n, r, K, method, trials, marker, color, lw, ls, label, folder=""):
    eps = read_eps_varying_r(n, r, K, method, ntrials, folder)
    eps_mean = np.mean(eps, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(K, eps_mean, marker=marker, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(K, eps_25, eps_75, color=color, lw=1, alpha=0.5)
    ax.fill_between(K, eps_10, eps_90, color=color, lw=1, alpha=0.3)
```

```python
K400 = [425, 450, 475, 500, 525, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000]
K800 = [825, 850, 875, 900, 925, 950, 975, 1000, 1050, 1100, 1150, 1200, 1250, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000]
K1600 = [1625, 1650, 1675, 1700, 1725, 1750, 1775, 1800, 1825, 1850, 1875, 1900, 1925, 1950, 1975, 2000]
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plot_k(ax, 2000, 400, K400, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=3, ls="-", label=r"$r = 400$")
plot_k(ax, 2000, 800, K800, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["bluegreen"], lw=3, ls="-", label=r"$r = 800$")
plot_k(ax, 2000, 1600, K1600, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["redpurple"], lw=3, ls="-", label=r"$r = 1600$")

x400 = np.linspace(405, 1000, 100)
ax.semilogy(x400, 5e-2 * (x400-400)**(0.25))
x800 = np.linspace(805, 1600, 100)
ax.semilogy(x800, 3.5e-2 * (x800-800)**(0.25))
x1600 = np.linspace(1605, 2000, 100)
ax.semilogy(x1600, 2.35e-2 * (x1600-1600)**(0.25))

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax.xaxis.set_major_locator(plt.FixedLocator([400, 800, 1200, 1600, 2000]))
# ax.set_ylim(1e-2, 1.0)
# ax.set_xlim(300, 1000)

# Titles
ax.set_title("Low-rank", fontsize=fontsize)
ax.set_xlabel(r"rank: $r$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)
```
