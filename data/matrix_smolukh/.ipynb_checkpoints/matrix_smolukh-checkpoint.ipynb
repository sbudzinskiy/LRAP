```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

### Load data

```python
data = np.loadtxt("./smolukh_1024.data")
m, n = data.shape
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
im = ax.imshow(data, origin='lower', cmap="cividis")

# Axes
# ax.set_aspect('equal')
ax.tick_params(labelsize=fontsize)
ax.xaxis.set_major_locator(plt.FixedLocator([0, 1023]))
ax.set_xticklabels([1, 1024])
ax.yaxis.set_major_locator(plt.FixedLocator([0, 1023]))
ax.set_yticklabels([1, 1024])

# Titles
# ax.set_title(f"Mass concentration at t = 6", fontsize=fontsize)

# Colorbar
cb = fig.colorbar(im, ax=ax)
cb.ax.tick_params(labelsize=fontsize)

# plt.savefig(f"./pics/smolukh.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```

### Singular values

```python
U, S, VT = np.linalg.svd(data, full_matrices=False)

smallest = []
for r in range(m):
    B = U[:, :r] @ np.diag(S[:r]) @ VT[:r, :]
    B[B > 0] = 0
    smallest.append(-np.amin(B))
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
ax.semilogy(np.arange(m)+1, S / S[0], lw=3, ls="-",  color=oi_colors["blue"],   label=r"$\sigma_r(X) / \sigma_1(X)$")
ax.semilogy(np.arange(m), smallest,   lw=3, ls="--", color=oi_colors["orange"], label=r"$|| \min(0, X_r) ||_{\max}$")

# Axes
ax.grid()
ax.tick_params(labelsize=fontsize)
ax.set_xlim(-2, 75)
ax.set_ylim(1e-15, 5e-0)
ax.yaxis.set_major_locator(plt.LogLocator(numticks=5))
ax.xaxis.set_major_locator(plt.FixedLocator([0, 25, 50, 75]))

# Titles
# ax.set_title(r"Absolute value of the smallest negative entry of $X_r$", fontsize=fontsize)
ax.set_xlabel(r"rank: $r$", fontsize=fontsize)

# Legend
ax.legend(loc='lower right', frameon=True, fontsize=fontsize, ncol=3)

# plt.savefig(f"./pics/smolukh_svs.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```

### AP for different LR projections

```python
def load_log(file):
    data = np.loadtxt(file, dtype=np.float64)

    n = data.shape[0] // 4
    data = np.reshape(data[:, 1], [n, 4])
    
    nnf = data[:, 0]
    nnc = data[:, 1]
    erf = data[:, 2]
    erc = data[:, 3]
    return nnf, nnc, erf, erc

def load_logs(folder, method_ap, method_lr, numexp, modifier=""):
    logs_nnf = []
    logs_nnc = []
    logs_erf = []
    logs_erc = []
    for seed in range(numexp):
        log = load_log(folder + f"{method_ap}_{method_lr}{modifier}/" + f"{seed+1}.log")
        logs_nnf.append(log[0])
        logs_nnc.append(log[1])
        logs_erf.append(log[2])
        logs_erc.append(log[3])
    
    return logs_nnf, logs_nnc, logs_erf, logs_erc

def log_stats(folder, method_ap, method_lr, numexp, modifier=""):
    logs_nnf, logs_nnc, logs_erf, logs_erc = load_logs(folder, method_ap, method_lr, numexp, modifier)
    
    stats = {}
    percentiles = [10, 25, 50, 75, 90]
    
    nnf = np.array(logs_nnf)
    for p in percentiles:
        stats[f"nnf_{p}"] = np.percentile(nnf, p, axis=0)
        
    nnc = np.array(logs_nnc)
    for p in percentiles:
        stats[f"nnc_{p}"] = np.percentile(nnc, p, axis=0)
        
    erf = np.array(logs_erf)
    erf = erf / erf[0,0]
    for p in percentiles:
        stats[f"erf_{p}"] = np.percentile(erf, p, axis=0)
        
    erc = np.array(logs_erc)
    for p in percentiles:
        stats[f"erC_{p}"] = np.percentile(erc, p, axis=0)
        
    return stats

def plot_stats_percentiles(ax, stats, what, label, color, ls, lw=3, log=True):
    n = stats[f"{what}_50"].shape[0]
    x = np.arange(n)
    
    if log:
        ax.semilogy(x, stats[f"{what}_50"], color=color, lw=lw, ls=ls, label=label)
    else:
        ax.plot(x, stats[f"{what}_50"], color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(x, stats[f"{what}_25"], stats[f"{what}_75"], color=color, lw=0, alpha=0.3)
    ax.fill_between(x, stats[f"{what}_10"], stats[f"{what}_90"], color=color, lw=0, alpha=0.15)
```

```python
ap_svd = log_stats("./", "ap", "svd", 1, modifier="")
ap_rsvd = log_stats("./", "ap", "rsvd", 10, modifier="_p2_q0")
ap_cross_warm = log_stats("./", "ap", "cross", 10, modifier="_warm")
ap_cross_cold = log_stats("./", "ap", "cross", 10, modifier="_cold")
ap_crossp_warm = log_stats("./", "ap", "crossp", 10, modifier="_k30_warm")
ap_crossp_cold = log_stats("./", "ap", "crossp", 10, modifier="_k30_cold")
```

```python
fig, ax = plt.subplots(figsize=(7, 4), constrained_layout=True)    
    
# Plot
plot_stats_percentiles(ax, ap_svd, "nnf", label=r"SVD", color=oi_colors["blue"], lw=3, ls="-")
plot_stats_percentiles(ax, ap_rsvd, "nnf", label=r"RSVD", color=oi_colors["orange"], lw=3, ls="-")
plot_stats_percentiles(ax, ap_cross_warm, "nnf", label=r"$\mathrm{VOL}_{w}$", color=oi_colors["vermilion"], lw=3, ls="-")
plot_stats_percentiles(ax, ap_cross_cold, "nnf", label=r"$\mathrm{VOL}_{c}$", color=oi_colors["skyblue"], lw=3, ls="-")
plot_stats_percentiles(ax, ap_crossp_warm, "nnf", label=r"$\mathrm{PVOL}_{w}$", color=oi_colors["redpurple"], lw=3, ls="-")
plot_stats_percentiles(ax, ap_crossp_cold, "nnf", label=r"$\mathrm{PVOL}_{c}$", color=oi_colors["bluegreen"], lw=3, ls="-")
ax.axhline(y=EPS, lw=3, color=oi_colors["black"], ls=":")

# Axes
ax.grid()
ax.tick_params(labelsize=fontsize)
ax.set_ylim(1e-12, 1e-1)
ax.set_xlim(-30, 1030)
ax.yaxis.set_major_locator(plt.LogLocator(numticks=4))
ax.xaxis.set_major_locator(plt.FixedLocator([0, 250, 500, 750, 1000]))

# Titles
ax.set_xlabel("iteration", fontsize=fontsize)
ax.set_ylabel(r"$|| \min(0, Y_k) ||_F$ / $|| X ||_F$", fontsize=fontsize)

# Legend
ax.legend(bbox_to_anchor=(1,1), frameon=False, fontsize=fontsize, ncol=1) #loc='lower center'

# plt.savefig(f"./pics/smolukh_ap.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/smolukh_ap.pdf", format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```
