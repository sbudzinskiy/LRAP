```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

### Fixed N

```python
def read_eps_varying_r(n, R, method, trials, folder=""):
    eps = np.zeros([len(R), len(trials)])
    for j in range(len(R)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{R[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_n(ax, n, R, method, trials, marker, color, lw, ls, label, folder="", minimum=False):
    eps = read_eps_varying_r(n, R, method, trials, folder)
    if minimum:
        eps_target = np.min(eps, axis=1)
    else:
#         eps_target = np.mean(eps, axis=1)
        eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    print(eps_50[1])
    
    ax.semilogy(R, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(R, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(R, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

# Joint 1x2 plot Fixed N [Magnum Opus]

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

R1600 = [20, 40, 80, 160, 320, 640]
x1600 = np.linspace(35, 640, 100)

##################################
###### Plot 1 
##################################

# Plot
plot_n(ax[0], 1600, R1600, "svd", np.arange(1, 11), marker="o", color=oi_colors["blue"], lw=3, ls="-", label=r"$n = 1600$", folder="matrix_randort/")

ax[0].semilogy(x1600, 19/x1600, color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1})$")
ax[0].semilogy(x1600, 6.75e-1/np.sqrt(x1600), color=oi_colors["orange"], lw=3, ls="-.", label=r"$O(r^{-1/2})$")

# Axes
ax[0].grid(axis='y')
ax[0].tick_params(labelsize=fontsize)
ax[0].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax[0].xaxis.set_major_locator(plt.FixedLocator([20, 320, 640]))
ax[0].set_ylim(1e-2, 1.0)

# Titles
ax[0].set_title("Random orthogonal", fontsize=fontsize)
ax[0].set_xlabel(r"rank: $r$", fontsize=fontsize)
ax[0].set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax[0].legend(loc='upper right', frameon=True, fontsize=fontsize, ncol=1)

##################################
###### Plot 2
##################################

# Plot
plot_n(ax[1], 1600, R1600, "svd", np.arange(1, 6), marker="o", color=oi_colors["blue"], lw=3, ls="-", label=r"$n = 1600$", folder="matrix_identity/", minimum=True)

ax[1].semilogy(x1600, 28/x1600, color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1})$")
ax[1].semilogy(x1600, 0.8/np.sqrt(x1600), color=oi_colors["orange"], lw=3, ls="-.", label=r"$O(r^{-1/2})$")

# Axes
ax[1].grid(axis='y')
ax[1].tick_params(labelsize=fontsize)
ax[1].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax[1].xaxis.set_major_locator(plt.FixedLocator([20, 320, 640]))
# ax.set_xlim(-50, 2000)
ax[1].set_ylim(1e-2, 1.0)

# Titles
ax[1].set_title("Identity", fontsize=fontsize)
ax[1].set_xlabel(r"rank: $r$", fontsize=fontsize)

# Legend
ax[1].legend(loc='upper right', frameon=True, fontsize=fontsize, ncol=1)

##################################
###### Layout
##################################
fig.tight_layout(h_pad=0, w_pad=1)

# plt.savefig(f"./pics/maxnorm_fixed_n.pdf", format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```

### Fixed R

```python
def read_eps_varying_n(r, N, method, trials, folder=""):
    eps = np.zeros([len(N), len(trials)])
    for j in range(len(N)):
        for i in range(len(trials)):
            with open(f'./{folder}n{N[j]}_r{r}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_r(ax, r, N, method, trials, marker, color, lw, ls, label, folder="", minimum=False):
    eps = read_eps_varying_n(r, N, method, trials, folder)
    if minimum:
        eps_target = np.min(eps, axis=1)
    else:
#         eps_target = np.mean(eps, axis=1)
        eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(N, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(N, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(N, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

# Joint 1x2 plot Fixed R [Magnum Opus]

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

# N40 = [100, 150, 200, 250, 300, 350, 400, 450, 500, 600, 700, 800, 900, 1000, 1500, 2000, 2500, 3000, 4000, 5000, 10000, 20000]
N40 = [100, 300, 500, 700, 1000, 1500, 2000, 2500, 3000, 4000, 5000, 10000, 20000]
y40 = np.linspace(50, 20000, 1000)

##################################
###### Plot 1 
##################################

# Plot
plot_r(ax[0], 40, N40, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=3, ls="-", label=r"$r = 40$", folder="matrix_randort/")
ax[0].semilogy(y40, 4.5e-1*np.log(y40)/np.sqrt(y40), color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(n^{-1/2} \log(n))$")

# Axes
ax[0].grid(axis='y')
ax[0].tick_params(labelsize=fontsize)
ax[0].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax[0].xaxis.set_major_locator(plt.FixedLocator([50, 10000, 20000]))
ax[0].set_ylim(1e-2, 1.0)

# Titles
ax[0].set_title("Random orthogonal", fontsize=fontsize)
ax[0].set_xlabel(r"size: $n$", fontsize=fontsize)
ax[0].set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax[0].legend(loc='upper right', frameon=True, fontsize=fontsize, ncol=1)

# ##################################
# ###### Plot 2
# ##################################

# # Plot
plot_r(ax[1], 40, N40, "rsvd_p5_q1", np.arange(1,6), marker="o", color=oi_colors["blue"], lw=3, ls="-", label=r"$r = 40$", folder="matrix_identity/", minimum=True)
ax[1].semilogy(y40, 6.5e-3*np.log(y40)**2, color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(\log^2(n))$")

# # Axes
ax[1].grid(axis='y')
ax[1].tick_params(labelsize=fontsize)
ax[1].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2]))
ax[1].xaxis.set_major_locator(plt.FixedLocator([50, 10000, 20000]))
ax[1].set_ylim(1e-2, 1.0)

# # Titles
ax[1].set_title("Identity", fontsize=fontsize)
ax[1].set_xlabel(r"size: $n$", fontsize=fontsize)

# # Legend
ax[1].legend(loc='lower right', frameon=True, fontsize=fontsize, ncol=1)

##################################
###### Layout
##################################
fig.tight_layout(h_pad=0, w_pad=1)

# plt.savefig(f"./pics/maxnorm_fixed_r.pdf", format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```
