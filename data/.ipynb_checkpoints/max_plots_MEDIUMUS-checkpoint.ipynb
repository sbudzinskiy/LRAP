```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import mark_inset, InsetPosition, zoomed_inset_axes
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 15
EPS = 1.1e-16
```

```python
def read_eps_varying_n(d, r, N, method, trials, folder=""):
    eps = np.zeros([len(N), len(trials)])
    for j in range(len(N)):
        for i in range(len(trials)):
            with open(f'./{folder}d{d}_n{N[j]}_r{r}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_r(ax, d, r, N, method, trials, marker, color, lw, ls, label, folder="", minimum=False):
    eps = read_eps_varying_n(d, r, N, method, trials, folder)
    if minimum:
        eps_target = np.min(eps, axis=1)
    else:
        eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(N, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(N, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(N, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

## Banded random uniform matrices

```python
def plot_norms(ax, K, nrms, marker, color, lw, ls, label):   
    nrms_target = np.percentile(nrms, 50, axis=1)
    nrms_50 = np.percentile(nrms, 50, axis=1)
    nrms_25 = np.percentile(nrms, 25, axis=1)
    nrms_75 = np.percentile(nrms, 75, axis=1)
    nrms_10 = np.percentile(nrms, 10, axis=1)
    nrms_90 = np.percentile(nrms, 90, axis=1)
    
    ax.semilogy(K, nrms_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(K, nrms_25, nrms_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(K, nrms_10, nrms_90, color=color, lw=0, alpha=0.3)
    
def read_eps_varying_b(n, r, B, method, trials, folder=""):
    eps = np.zeros([len(B), len(trials)])
    for j in range(len(B)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{r}_b{B[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_b(ax, n, r, B, method, trials, marker, color, lw, ls, label, folder=""):
    eps = read_eps_varying_b(n, r, B, method, trials, folder)
    eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(B, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(B, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(B, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
fig.tight_layout(h_pad=0, w_pad=-1.5)

######################################
###### Plot 1: Norms
######################################

npzfile = np.load('./matrix_banded_uniform/nrms_K.npz')
nrms = npzfile['nrms']
B = npzfile['K']

# Plot
plot_norms(ax[0], B, nrms, marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"spectral norm")
ax[0].semilogy(B, np.sqrt(1*B), color=oi_colors["vermilion"], lw=2, ls="--", label=r"$\sqrt{b}$")
ax[0].semilogy(B, np.sqrt(3*B), color=oi_colors["orange"], lw=2, ls="-.", label=r"$\sqrt{3b}$")

# Axes
ax[0].grid(axis='y')
ax[0].tick_params(labelsize=fontsize)
ax[0].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e1, 1e2, 1e3]))
ax[0].xaxis.set_major_locator(plt.FixedLocator([1, 500, 1000, 1500, 2000]))
ax[0].set_ylim(0.9, 1e2)

# Titles
ax[0].set_xlabel(r"band width: $b$", fontsize=fontsize)

# Legend
ax[0].legend(loc='lower right', frameon=False, fontsize=fontsize-1, ncol=3)


######################################
###### Plot 2: Fixed N,R
######################################

B2000 = [1, 10, 20, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000]

# Plot
plot_b(ax[1], 2000, 800, B2000, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"$r = 800$",
       folder="matrix_banded_uniform/")
plot_b(ax[1], 2000, 1600, B2000, "rsvd_p5_q1", np.arange(1,11), marker="s", color=oi_colors["bluegreen"], lw=1, ls="-", label=r"$r = 1600$",
       folder="matrix_banded_uniform/")

plot_norms(ax[1], B, 8e-3*nrms, marker="", color=oi_colors["vermilion"], lw=2, ls="--", label=r"$O$(spectral norm)")
plot_norms(ax[1], B, 1.4e-3*nrms, marker="", color=oi_colors["vermilion"], lw=2, ls="--", label="")

# Axes
ax[1].grid(axis='y')
ax[1].tick_params(labelsize=fontsize)
ax[1].tick_params(axis='y', which='both', left=False, right=True, labelleft=False, labelright=True)
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2, 1e-3]))
ax[1].xaxis.set_major_locator(plt.FixedLocator([1, 500, 1000, 1500, 2000]))
ax[1].set_ylim(1e-3, 1.2)

# Titles
ax[1].set_xlabel(r"band width: $b$", fontsize=fontsize)
ax[1].set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize, rotation=-90, va="bottom")

# Legend
ax[1].legend(loc='lower right', frameon=False, fontsize=fontsize-1, ncol=3)

# plt.savefig(f"./pics/minorum/banded_uniform.pdf", format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```

## Products of random matrices with orthonormal columns

```python
def read_eps_varying_k(n, r, K, method, trials, folder=""):
    eps = np.zeros([len(K), len(trials)])
    for j in range(len(K)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{r}_k{K[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_k(ax, n, r, K, method, trials, marker, color, lw, ls, label, folder=""):
    eps = read_eps_varying_k(n, r, K, method, trials, folder)
    eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(K, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(K, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(K, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```
