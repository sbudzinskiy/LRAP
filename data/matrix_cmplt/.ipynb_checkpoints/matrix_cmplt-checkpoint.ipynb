```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

```python
OS = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]
```

```python
def measure_success_rate(reg, tol):
    success = np.zeros([len(OS), 20])
    for i in range(13):
        for seed in range(20):
            data = np.loadtxt(f"./os{OS[i]}_{reg}/{seed+1}.log")
            success[i, seed] = np.any(data[:, 1] < tol)
    success_rate = np.mean(success, axis=1)
    return success_rate
```

```python
noreg = measure_success_rate("nap0_rp_cross", 1e-6)
reg = measure_success_rate("nap1_ap_cross", 1e-6)
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plt.plot(OS, noreg, marker="o", color=oi_colors["blue"], lw=3, label="no reg.")
plt.plot(OS, reg, marker="o", color=oi_colors["orange"], lw=3, label=r"AP-$\mathrm{VOL}_{w}$")

# Axes
ax.grid()
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.MaxNLocator(5))

# Titles
# ax.set_title(r"Absolute value of the smallest negative entry of $X_r$", fontsize=fontsize)
ax.set_xlabel(r"oversampling: $\gamma$", fontsize=fontsize)
ax.set_ylabel("frequency of success", fontsize=fontsize)

# Legend
ax.legend(loc='lower right', frameon=True, fontsize=fontsize, ncol=1)

# plt.savefig(f"./pics/completion.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/completion.pdf", format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```
