### Spectral norm of banded uniform matrix

```python
def make_banded(X, b):
    m, n = X.shape
    XX = np.copy(X)
    for j in range(n):
        for i in range(j-b+1):
            XX[i,j] = 0
        for i in range(j+b,m):
            XX[i,j] = 0
    return XX

def norm_banded_random(n, k, ntrials):
    norms = np.zeros(ntrials, dtype='float')
    for i in range(ntrials):
        X = np.random.default_rng().uniform(-1,1,[n, n])
        XX = make_banded(X, k)
        norms[i] = np.linalg.norm(XX, 2)
    return norms

def norm_banded_random_stats(n, K, ntrials):
    nrms = np.zeros([len(K),ntrials], dtype='float')
    for i in range(len(K)):
        nrms[i] = norm_banded_random(n, K[i], ntrials)
    return nrms

def plot_norms(ax, K, nrms, marker, color, lw, ls, label):   
    nrms_target = np.percentile(nrms, 50, axis=1)
    nrms_50 = np.percentile(nrms, 50, axis=1)
    nrms_25 = np.percentile(nrms, 25, axis=1)
    nrms_75 = np.percentile(nrms, 75, axis=1)
    nrms_10 = np.percentile(nrms, 10, axis=1)
    nrms_90 = np.percentile(nrms, 90, axis=1)
    
    ax.semilogy(K, nrms_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(K, nrms_25, nrms_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(K, nrms_10, nrms_90, color=color, lw=0, alpha=0.3)
```

### Fixed n and r

```python
def read_eps_varying_r(n, r, B, method, trials, folder=""):
    eps = np.zeros([len(B), len(trials)])
    for j in range(len(B)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{r}_b{B[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_b(ax, n, r, B, method, trials, marker, color, lw, ls, label, folder=""):
    eps = read_eps_varying_r(n, r, B, method, trials, folder)
    eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(B, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(B, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(B, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
B2000 = [1, 10, 20, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800, 1900, 2000]
```

```python
fig, ax = plt.subplots(figsize=(8, 4), constrained_layout=True)    

# Plot
# plot_b(ax, 2000, 400, B2000, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"$n = 2000, r = 400$")
plot_b(ax, 2000, 800, B2000, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"$n = 2000, r = 800$")
plot_b(ax, 2000, 1600, B2000, "rsvd_p5_q1", np.arange(1,11), marker="s", color=oi_colors["bluegreen"], lw=1, ls="-", label=r"$n = 2000, r = 1600$")

x400 = np.linspace(1, 2000, 100)
ax.semilogy(x400, 1.2e-2*np.sqrt(x400)/np.log(10+x400), color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1})$")
# ax.semilogy(x400, 7.2e-2*np.sqrt(x400)/np.log(20+x400), color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1})$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2, 1e-3]))
ax.xaxis.set_major_locator(plt.FixedLocator([1, 500, 1000, 1500, 2000]))
ax.set_ylim(1e-3, 1.2)

# Titles
ax.set_xlabel(r"band width: $b$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='lower right', frameon=True, fontsize=fontsize, ncol=1)
```
