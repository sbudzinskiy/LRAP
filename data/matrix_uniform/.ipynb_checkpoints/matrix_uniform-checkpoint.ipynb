```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import mark_inset, InsetPosition, zoomed_inset_axes
%matplotlib inline

oi_colors = {
    "orange"    : "#E69F00", 
    "skyblue"   : "#56B4E9", 
    "bluegreen" : "#009E73", 
    "yellow"    : "#F0E442", 
    "blue"      : "#0072B2", 
    "vermilion" : "#D55E00", 
    "redpurple" : "#CC79A7",
    "black"     : "#000000"}

fontsize = 14
EPS = 1.1e-16
```

### Fixed n

```python
def read_eps_varying_r(n, R, method, trials, folder=""):
    eps = np.zeros([len(R), len(trials)])
    for j in range(len(R)):
        for i in range(len(trials)):
            with open(f'./{folder}n{n}_r{R[j]}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_n(ax, n, R, method, trials, marker, color, lw, ls, label, folder="", minimum=False):
    eps = read_eps_varying_r(n, R, method, trials, folder)
    eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(R, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(R, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(R, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
R5000 = [50, 100, 200, 400, 600, 800, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 4700, 4900]
R3000 = [50, 100, 200, 400, 600, 800, 1000, 1500, 2000, 2500, 2700, 2900]
R1500 = [50, 100, 200, 400, 600, 800, 1000, 1200, 1400]
```

```python
fig, ax = plt.subplots(figsize=(8, 4), constrained_layout=True)    

# Plot
plot_n(ax, 5000, R5000, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"$n = 5000$")
plot_n(ax, 3000, R3000, "rsvd_p5_q1", np.arange(1,11), marker="s", color=oi_colors["bluegreen"], lw=1, ls="-", label=r"$n = 3000$")
plot_n(ax, 1500, R1500, "rsvd_p5_q1", np.arange(1,11), marker="v", color=oi_colors["redpurple"], lw=1, ls="-", label=r"$n = 1500$")

x5000 = np.linspace(800, 5000, 100)
ax.plot(x5000, 3e-1*(5000-x5000)/x5000, color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(r^{-1} \cdot (n - r))$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2, 1e-3]))
ax.xaxis.set_major_locator(plt.FixedLocator([50, 1500, 3000, 5000]))
ax.set_ylim(1e-3, 1.2)

# Titles
ax.set_xlabel(r"rank: $r$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='lower left', frameon=True, fontsize=fontsize-1, ncol=2)
```

### Fixed r

```python
def read_eps_varying_n(r, N, method, trials, folder=""):
    eps = np.zeros([len(N), len(trials)])
    for j in range(len(N)):
        for i in range(len(trials)):
            with open(f'./{folder}n{N[j]}_r{r}_{method}/{trials[i]}.log') as f:
                for line in f:
                    pass
                eps[j,i] = float(line)
    return eps

def plot_r(ax, r, N, method, trials, marker, color, lw, ls, label, folder="", minimum=False):
    eps = read_eps_varying_n(r, N, method, trials, folder)
    eps_target = np.percentile(eps, 50, axis=1)
    eps_50 = np.percentile(eps, 50, axis=1)
    eps_25 = np.percentile(eps, 25, axis=1)
    eps_75 = np.percentile(eps, 75, axis=1)
    eps_10 = np.percentile(eps, 10, axis=1)
    eps_90 = np.percentile(eps, 90, axis=1)
    
    ax.semilogy(N, eps_target, marker=marker, markersize=5, color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(N, eps_25, eps_75, color=color, lw=0, alpha=0.5)
    ax.fill_between(N, eps_10, eps_90, color=color, lw=0, alpha=0.3)
```

```python
N50  =  [75, 100, 125, 150, 200, 250, 300, 350, 400, 500, 600, 700, 800, 900, 1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 20000]
N100  = [125, 150, 200, 250, 300, 350, 400, 450, 500, 600, 700, 800, 900, 1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 20000]
N200  = [225, 250, 300, 350, 400, 450, 500, 600, 700, 800, 900, 1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 20000]
N400  = [425, 450, 500, 550, 600, 650, 700, 800, 900, 1000, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 20000]
N800  = [825, 850, 900, 950, 1000, 1100, 1200, 1400, 1600, 1800, 2000, 2500, 3000, 4000, 5000, 7500, 10000, 20000]
```

```python
fig, ax = plt.subplots(figsize=(8, 4), constrained_layout=True)    

# Plot
plot_r(ax, 50, N50, "rsvd_p5_q1", np.arange(1,11), marker="o", color=oi_colors["blue"], lw=1, ls="-", label=r"$r = 50$")
# plot_r(ax, 100, N100, "rsvd_p5_q1", 1+np.arange(5), marker="s", color=oi_colors["bluegreen"], lw=3, ls="-", label=r"$r = 100$")
plot_r(ax, 200, N200, "rsvd_p5_q1", np.arange(1,11), marker="s", color=oi_colors["bluegreen"], lw=1, ls="-", label=r"$r = 200$")
# plot_r(ax, 400, N400, "rsvd_p5_q1", 1+np.arange(5), marker="s", color=oi_colors["orange"], lw=3, ls="-", label=r"$r = 400$")
plot_r(ax, 800, N800, "rsvd_p5_q1", np.arange(1,11), marker="v", color=oi_colors["redpurple"], lw=1, ls="-", label=r"$r = 800$")

# y800 = np.linspace(1600, 20000, 1000)
# ax.semilogy(y800, 1e-5*np.log(y800 - 800)**3 * y800**0.5, color=oi_colors["vermilion"], lw=3, ls="--", label=r"$O(\log^3(n-r))$")

# Axes
ax.grid(axis='y')
ax.tick_params(labelsize=fontsize)
ax.yaxis.set_major_locator(plt.FixedLocator([1e-0, 1e-1, 1e-2, 1e-3]))
ax.xaxis.set_major_locator(plt.FixedLocator([50, 10000, 20000]))
ax.set_ylim(1e-3, 1.2)

# Titles
ax.set_xlabel(r"size: $n$", fontsize=fontsize)
ax.set_ylabel(r"error bound: $\varepsilon_{+}$", fontsize=fontsize)

# Legend
ax.legend(loc='lower center', frameon=True, fontsize=fontsize-1, ncol=4)

# Inset plot
axin = plt.axes([0,0,1,1])
ip = InsetPosition(ax, [0.35,0.25,0.6,0.3])
axin.set_axes_locator(ip)

plot_r(axin, 200, N200[:10], "rsvd_p5_q1", np.arange(1,11), marker="s", color=oi_colors["bluegreen"], lw=1, ls="-", label="")

x200 = np.linspace(210, 800, 100)
axin.semilogy(x200, 6e-3 * (x200-200)**0.75, color=oi_colors["orange"], lw=2, ls="-.", label=r"$O((n-r)^{3/4})$")

# Axes
axin.grid(axis='y')
axin.tick_params(labelsize=fontsize-3)
axin.yaxis.set_major_locator(plt.FixedLocator([]))
axin.yaxis.set_minor_locator(plt.FixedLocator([]))
axin.xaxis.set_major_locator(plt.FixedLocator([200, 800]))
axin.set_ylim(1e-2, 1.0)

# Legend
axin.legend(loc='lower right', frameon=True, fontsize=fontsize-1, ncol=4)

# ax.set_rasterized(True)
# plt.savefig(f"./pics/randort_varying_n_rsvd.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/randort_varying_n_rsvd.eps", format='eps', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/mettx_identity_n.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```
