### Plot negative elements in the approximation

```python
def plot_negative(ax, f, x, label, color, alpha, lw, ls):
    n = len(x)
    pos = f > 0
    neg = f < 0
    
    ax.semilogy(x, np.abs(f), label=label, color=color, lw=lw, ls=ls)
    
    left_pos = 0
    while any(neg[left_pos:]) and left_pos < n-1:
        
        left_neg = left_pos + np.where(neg[left_pos:])[0][0]
        if (any(pos[left_neg:])):
            dx = np.where(pos[left_neg:])[0][0]
        else:
            dx = n - left_neg - 1
        left_pos = left_neg + dx
        
        if (left_neg == 0):
            f_left = f[left_pos]
        else:
            f_left = f[left_neg-1]
        ax.axvspan(x[left_neg], x[left_pos], alpha=alpha, color=color)
    return
```

### Plot convergence

```python
def load_log(file):
    data = np.loadtxt(file, dtype=np.float64)

    n = data.shape[0] // 4
    data = np.reshape(data[:, 1], [n, 4])
    
    nnf = data[:, 0]
    nnc = data[:, 1]
    erf = data[:, 2]
    erc = data[:, 3]
    return nnf, nnc, erf, erc

def load_logs(folder, method_ap, method_lr, numexp, modifier=""):
    logs_nnf = []
    logs_nnc = []
    logs_erf = []
    logs_erc = []
    for seed in range(numexp):
        log = load_log(folder + f"{method_ap}_{method_lr}{modifier}/" + f"{seed+1}.log")
        logs_nnf.append(log[0])
        logs_nnc.append(log[1])
        logs_erf.append(log[2])
        logs_erc.append(log[3])
    
    return logs_nnf, logs_nnc, logs_erf, logs_erc

def log_stats(folder, method_ap, method_lr, numexp, modifier=""):
    logs_nnf, logs_nnc, logs_erf, logs_erc = load_logs(folder, method_ap, method_lr, numexp, modifier)
    
    stats = {}
    percentiles = [10, 25, 50, 75, 90]
    
    nnf = np.array(logs_nnf)
    for p in percentiles:
        stats[f"nnf_{p}"] = np.percentile(nnf, p, axis=0)
        
    nnc = np.array(logs_nnc)
    for p in percentiles:
        stats[f"nnc_{p}"] = np.percentile(nnc, p, axis=0)
        
    erf = np.array(logs_erf)
    erf = erf / erf[0,0]
    for p in percentiles:
        stats[f"erf_{p}"] = np.percentile(erf, p, axis=0)
        
    erc = np.array(logs_erc)
    for p in percentiles:
        stats[f"erC_{p}"] = np.percentile(erc, p, axis=0)
        
    return stats

def plot_stats_percentiles(ax, stats, what, label, color, ls, lw=3, log=True, n=-1):
    if n == -1:
        n = stats[f"{what}_50"].shape[0]
    x = np.arange(n)
    
    if log:
        ax.semilogy(x, stats[f"{what}_50"][:n], color=color, lw=lw, ls=ls, label=label)
    else:
        ax.plot(x, stats[f"{what}_50"][:n], color=color, lw=lw, ls=ls, label=label)
    ax.fill_between(x, stats[f"{what}_25"][:n], stats[f"{what}_75"][:n], color=color, lw=0, alpha=0.3)
    ax.fill_between(x, stats[f"{what}_10"][:n], stats[f"{what}_90"][:n], color=color, lw=0, alpha=0.15)
```

```python
ap = log_stats("./", "ap", "ttsvd", 1, modifier="")
rp = log_stats("./", "rp", "ttsvd", 1, modifier="")
sp = log_stats("./", "sp_0.5", "ttsvd", 1, modifier="")
ip = log_stats("./", "ip_0.75", "ttsvd", 1, modifier="")
```

```python
fig, ax = plt.subplots(figsize=(6, 4), constrained_layout=True)    

# Plot
plot_stats_percentiles(ax, ap, "nnf", label=r"AP", color=oi_colors["blue"], lw=3, ls="-", n=51)
plot_stats_percentiles(ax, rp, "nnf", label=r"RP", color=oi_colors["orange"], lw=3, ls="--", n=51)
plot_stats_percentiles(ax, sp, "nnf", label=r"SP$(1/2)$", color=oi_colors["vermilion"], lw=3, ls="-.", n=51)
plot_stats_percentiles(ax, ip, "nnf", label=r"IP$(3/4)$", color=oi_colors["bluegreen"], lw=3, ls=(0, (3, 1, 1, 1, 1, 1)), n=51)
ax.axhline(y=EPS, lw=3, color=oi_colors["black"], ls=":")

# Axes
ax.grid()
ax.tick_params(labelsize=fontsize)
ax.set_ylim(1e-17, 1e-1)
# ax.set_xlim(-30, 1030)
ax.set_xlim(-1, 51)
ax.yaxis.set_major_locator(plt.LogLocator(numticks=4))
ax.xaxis.set_major_locator(plt.FixedLocator([0, 10, 20, 30, 40, 50]))
# ax.xaxis.set_major_locator(plt.FixedLocator([0, 250, 500, 750, 1000]))

# Titles
ax.set_xlabel("iteration", fontsize=fontsize)
ax.set_ylabel(r"$|| \min(0, Y_k) ||_F$ / $|| f ||_F$", fontsize=fontsize)

# Legend
ax.legend(loc='upper right', frameon=True, fontsize=fontsize, ncol=2)
# 
# plt.savefig(f"./pics/mixture_ap_convergence_1000.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
# plt.savefig(f"./pics/mixture_ap_convergence_50_extended.png", format='png', dpi=300, bbox_inches='tight', pad_inches=0)
```
